# 3. Create Four Variations for each character

1. First Variation: Shift Up

```python
#define function for shift up of the three letters
def shift_up(matrix):
    shifted_matrix = np.zeros_like(matrix)
    shifted_matrix[:-1] = matrix[1:]
    return shifted_matrix
```

```python
#function to display matrix that has been shifted
def display_matrix(matrix, title="Matrix"):
    plt.imshow(matrix, interpolation='nearest')
    plt.title(title)
    plt.show()
```

Shift up of the letter J

```python
#shift of letter J
J_shift_up = shift_up(J)
display_matrix(J_shift_up, "Shifted J")
```

Shift up of the letter U

```python
#shift up of letter U
U_shift_up = shift_up(U)
display_matrix(U_shift_up, "Shifted U")
```

Shift up of the letter I

```python
#shift of letter I
I_shift_up = shift_up(I)
display_matrix(I_shift_up, "Shifted I")
```

2. Second Variation: Blurred Matrix

```python
#define function for to blur to the three letters

def letter_blur(letter):
    blurring = np.zeros_like(letter, dtype=float)

    for i in range(letter.shape[0]):

        for j in range(letter.shape[1]):
            neigh = []

            for ni in range(max(0, i-1), min(i+2, letter.shape[0])):

                for nj in range(max(0, j-1), min(j+2, letter.shape[1])):
                    let = letter[ni, nj]
                    neigh.append(let)
            neigh_mean= np.mean(neigh)
            blurring[i, j] = neigh_mean
    return blurring
```

Blurred version of the letter J

```python
#Blurred matrix for the letter J
J_blurred = letter_blur(J_shift_up)
display_matrix(J_blurred, "Blurred J")
```

Blurred version of the letter U

```python
#Blurred matrix for the letter U
U_blurred = letter_blur(U_shift_up)
display_matrix(U_blurred, "Blurred U")
```

Blurred version of the letter I

```python
#Blurred matrix for the letter I
I_blurred = letter_blur(I_shift_up)
display_matrix(I_blurred, "Blurred I")
```

3. Third Variation: Noise Matrix

```python
#define function for to add noise to the three letters
#gaussian noise

def apply_noise(matrix, scale):
  matrix = np.asarray(matrix)
  noise = np.random.normal(loc=1.0, scale= scale, size=matrix.shape)
  noisy_matrix = matrix * noise
  return noisy_matrix
```

Noise version of the letter J

```python
#Noise matrix for the letter J
J_noise = apply_noise(J, 0.1)
display_matrix(J_noise, "Noisy J")
```

```python
# Function to average variations for each letter
def average_variations(*variations):
    return np.mean(np.array(variations), axis=0)

# Averaging variations for J, U, and I
J_avg = average_variations(J_shift_up, J_blurred, J_noise, J_noise_and_blur)
U_avg = average_variations(U_shift_up, U_blurred, U_noise, U_noise_and_blur)
I_avg = average_variations(I_shift_up, I_blurred, I_noise, I_noise_and_blur)

letters_avg = [J_avg, U_avg, I_avg]

# Centering (normalization) function
def center_matrix(matrix):
    centered = np.copy(matrix)
    for i in range(len(centered)):
        centered[i] = centered[i] - np.mean(centered[i])  # Subtracting the mean for normalization
    return centered

# Center all the average variations
J_avg_centered = center_matrix(J_avg)
U_avg_centered = center_matrix(U_avg)
I_avg_centered = center_matrix(I_avg)

# Compile NN1 matrix based on the centered average variations (NN1 is now 3x25 matrix)
NN1 = np.array([J_avg_centered.flatten(), U_avg_centered.flatten(), I_avg_centered.flatten()])

# Function to test a letter against NN1
def recognize_letter(input_matrix, weight_matrix):
    # Normalize input (similar to how we normalized average variations)
    centered_input = center_matrix(input_matrix).flatten()

    # Calculate dot products between the input and each row in NN1
    outputs = np.matmul(weight_matrix, centered_input)

    # Avoid dividing by zero by adding a small epsilon value if the norm is zero
    norm = np.linalg.norm(outputs)
    if norm == 0:
        normalized_outputs = outputs  # If the norm is zero, don't normalize to avoid NaNs
    else:
        normalized_outputs = outputs / norm

    return np.argmax(normalized_outputs), normalized_outputs  # Return the index of the best match and scores

# Letters and variations for testing
Letters = ['J', 'U', 'I']
J_Var = [J_shift_up, J_blurred, J_noise, J_noise_and_blur]
U_Var = [U_shift_up, U_blurred, U_noise, U_noise_and_blur]
I_Var = [I_shift_up, I_blurred, I_noise, I_noise_and_blur]

Variations = [J_Var, U_Var, I_Var]

# Test recognition on all character variations
for letter_idx, letter_variations in enumerate(Variations):
    print(f"\nTesting variations for letter {Letters[letter_idx]}:")
    for variation_idx, variation in enumerate(letter_variations):
        predicted_idx, scores = recognize_letter(variation, NN1)
        predicted_letter = Letters[predicted_idx]
        print(f"  Variation {variation_idx+1}: Recognized as {predicted_letter} (Scores: {scores})")

# Final predictions based on average characters
print("\nTesting recognition on the averaged characters:")
for letter_idx, avg_letter in enumerate(letters_avg):
    predicted_idx, scores = recognize_letter(avg_letter, NN1)
    predicted_letter = Letters[predicted_idx]
    print(f"Averaged {Letters[letter_idx]}: Recognized as {predicted_letter} (Scores: {scores})")
```
